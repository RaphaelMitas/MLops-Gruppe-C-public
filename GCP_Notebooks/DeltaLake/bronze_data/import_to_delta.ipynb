```python
df_pageinated = pd.read_json('/home/mlops/project/Ticketmaster/ticketmaster_data.json')
df_pageinated = df_pageinated.T
df_pageinated
```

```python
events_list = list()

for index, page in df_pageinated.iterrows():
    events_list += page['_embedded']['events']
events_df = pd.DataFrame(events_list)

#events_df = events_df.drop('seatmap', axis=1)
#events_df = events_df.drop('promoters', axis=1)
#events_df = events_df.drop('promoter', axis=1)

events_df
```

```python
events_df_compact = events_df.iloc[
    :,
    events_df.columns.get_indexer(
        ['name', 'id', 'locale', 'distance',
         'units', 'dates', 'classifications',
         'priceRanges', '_links', '_embedded']
    )
                    ]


#PRICE RANGES
events_df_compact['price_min'] = ""
events_df_compact['price_max'] = ""
for index, row in events_df_compact.iterrows():
    standard_price_min = -1.0
    standard_price_max = -1.0
    if isinstance(row.priceRanges, list):
        for price in row.priceRanges:
            if isinstance(price, dict) and price['type'] == 'standard':
                standard_price_min = price['min']
                standard_price_max = price['max']
    events_df_compact.iloc[index, events_df_compact.columns.get_loc('price_min')] = standard_price_min
    events_df_compact.iloc[index, events_df_compact.columns.get_loc('price_max')] = standard_price_max
events_df_compact = events_df_compact.drop('priceRanges', axis=1)

spark_df = spark.createDataFrame(events_df_compact)
events_df_compact.head()
```

```python
spark_df.write.format("delta").mode("overwrite").option("mergeSchema", "true").save("/home/mlops/project/DeltaLake/bronze_data/ticketmaster_table")
```
