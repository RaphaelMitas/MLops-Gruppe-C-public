```python
import os
os.environ["PYARROW_IGNORE_TIMEZONE"] = "1"
import pyspark
from delta import *

builder = pyspark.sql.SparkSession.builder.appName("MyApp") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")

spark = configure_spark_with_delta_pip(builder).getOrCreate()

```

```python
import pyspark.pandas as ps


pandas_df = ps.read_csv('/home/mlops/project/DeltaLake/bronze_data/genre_map.csv', ';')

pandas_df
```

```python

spark_df = pandas_df.to_spark()


spark_df.show()
```

```python

spark_df.write.format("delta").mode("overwrite").save("/home/mlops/project/DeltaLake/bronze_data/genre_map_table")
```

```python

genre_map = spark.read.format('delta').load('/home/mlops/project/DeltaLake/bronze_data/genre_map_table')
genre_map.show()
```
